```python
from thermogravimetricAnalysis import TGA
tga = TGA()
```

```python
tga.read('angelim', 5, '../examples/assets/T5/angelim.txt')
tga.read('angelim', 10, '../examples/assets/T10/angelim.txt')
tga.read('angelim', 15, '../examples/assets/T15/angelim.txt')
tga.read('angelim', 20, '../examples/assets/T20/angelim.txt')
tga.read('ipe_champagne', 5, '../examples/assets/T5/ipe_champagne.txt')
tga.read('ipe_champagne', 10, '../examples/assets/T10/ipe_champagne.txt')
tga.read('ipe_champagne', 15, '../examples/assets/T15/ipe_champagne.txt')
tga.read('ipe_champagne', 20, '../examples/assets/T20/ipe_champagne.txt')
```

```python
tga.massLossAndConversions()
```

```python
tga.dtgFull()
```

```python
tga.plotTGA('angelim');
```

```python
tga.plotTGA('ipe_champagne');
```

```python
tga.plotDTG('angelim', False)
```

```python
tga.plotDTG('ipe_champagne', False)
```

```python
for h in [5, 10, 15, 20]:
    id_to_drop = tga.data['angelim'][h][tga.data['angelim'][h]['dtg']>0.3] .index.to_list()
    tga.removeData(id_to_drop, 'angelim', h)

    id_to_drop2 = tga.data['angelim'][h][(tga.data['angelim'][h]['dtg']<=0.3) & (tga.data['angelim'][h]['temperature']<=200)] .index.to_list()
    tga.removeData(id_to_drop2, 'angelim', h)

    id_to_drop = tga.data['ipe_champagne'][h][tga.data['ipe_champagne'][h]['dtg']>0.3] .index.to_list()
    tga.removeData(id_to_drop, 'ipe_champagne', h)

    id_to_drop2 = tga.data['ipe_champagne'][h][(tga.data['ipe_champagne'][h]['dtg']<=0.3) & (tga.data['ipe_champagne'][h]['temperature']<=200)] .index.to_list()
    tga.removeData(id_to_drop2, 'ipe_champagne', h)
```

```python
tga.plotDTG('ipe_champagne', False)
```
